```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("CC_GENERAL.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
df_copy=df.copy()
```

```python
del df_copy['CUST_ID']
```

```python
df_copy.dropna(axis=0, inplace=True)
df_copy.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
StSc=StandardScaler()
df_copy_transfm=StSc.fit_transform(df_copy)
df_copy_transfm_df = pd.DataFrame (df_copy_transfm, columns = ['BALANCE','BALANCE_FREQUENCY','PURCHASES','ONEOFF_PURCHASES','INSTALLMENTS_PURCHASES','CASH_ADVANCE','PURCHASES_FREQUENCY','ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY','CASH_ADVANCE_FREQUENCY','CASH_ADVANCE_TRX','PURCHASES_TRX','CREDIT_LIMIT','PAYMENTS','MINIMUM_PAYMENTS','PRC_FULL_PAYMENT','TENURE'])
df_copy_transfm_df.head()
```

```python
dfcopy_for_procss=df_copy_transfm_df[['BALANCE','PURCHASES','CASH_ADVANCE','CREDIT_LIMIT','PAYMENTS']]
```

```python
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
```

```python
plot_correlation_map(df_copy_transfm_df)
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(dfcopy_for_procss)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative.head(10)
```

```python
unique, counts =np.unique(clust_labels, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=2, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df_copy_transfm_df)
```

```python
Sum_of_squared_distances=[]
K=range(1,15)
for k in K:
    Km=KMeans(n_clusters=k)
    Km=Km.fit(df_copy_transfm_df)
    Sum_of_squared_distances.append(Km.inertia_)
```

```python
plt.plot(K,Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

According to the Elbow Method, the best choice for the Value of K is 2
